```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Files to Load 
income1819_to_load = "Resources/income_1819.csv"
income1716_to_load = "Resources/income_1716.csv"


# Read the Income Data
income1819_df = pd.read_csv(income1819_to_load)
income1716_df = pd.read_csv(income1716_to_load)
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(income1819_df, income1716_df, how="left", on=["GEO_ID", "GEO_ID"])

# Display the data table for preview
merged_df.head(5)
```

```python
# Export the CSV
merged_df.to_csv('income_1.csv', index = False)
```
