```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Files to Load 
income1514_to_load = "Resources/income_1514.csv"
income_1_to_load = "Resources/income_1.csv"


# Read the Income Data
income1514_df = pd.read_csv(income1514_to_load)
income_1_df = pd.read_csv(income_1_to_load)
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(income1514_df, income_1_df, how="left", on=["GEO_ID", "GEO_ID"])

# Display the data table for preview
merged_df.head(5)
```

```python
# Export the CSV
merged_df.to_csv('income_final.csv', index = False)
```
