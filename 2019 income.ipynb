```python
# Import dependencies
import pandas as pd
from functools import reduce
```

```python
# Load in CSV as a DataFrame
income_df = pd.read_csv('Resources/ACSST5Y2019_income_data.csv')
income_df.head(10)
```

```python
# Drop null values
income_df.dropna()
```

```python
# Keep columns: income less than 50K
income_df50 = income_df.filter(['GEO_ID', 'Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!Less than $5,000', 'Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$5,000 to $9,999','Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$10,000 to $14,999', 'Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$15,000 to $19,999', 'Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$20,000 to $24,999', 'Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$25,000 to $34,999', 'Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$35,000 to $49,999'])
income_df50
```

```python
# Keep columns: income greater than 50K, less than 75K
income_df75  = income_df.filter(['GEO_ID','Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$50,000 to $74,999'])
income_df75
```

```python
# Keep columns: income greater than 75K, less than 100K
income_df_greaterthan75  = income_df.filter(['GEO_ID','Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$75,000 to $99,999'])
income_df_greaterthan75
```

```python
# Keep columns: income greater than 100K, less than 150K
income_df_greaterthan100  = income_df.filter(['GEO_ID','Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$100,000 to $149,999'])
income_df_greaterthan100
```

```python
# Keep columns: income greater than 150K
income_df_greaterthan150  = income_df.filter(['GEO_ID','Estimate!!Occupied housing units!!Occupied housing units!!HOUSEHOLD INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)!!$150,000 or more'])
income_df_greaterthan150
```

```python
# Keep columns: Household Median and Mean income and percent allocated
#MedMean_df  = income_df.filter(['GEO_ID', 'Households!!Estimate!!Median income (dollars)','Households!!Estimate!!Mean income (dollars)','Households!!Estimate!!PERCENT IMPUTED!!Household income in the past 12 months'])
#MedMean_df
```

```python
#Merge dataframes
dfs = [income_df50, income_df75, income_df_greaterthan75, income_df_greaterthan100, income_df_greaterthan150]
income_df = reduce(lambda left,right: pd.merge(left,right,on=['GEO_ID'],
                                            how='outer'), dfs)
income_df.head()
```

```python
# Export the CSV
income_df.to_csv('income_2019.csv', index = False)
```
