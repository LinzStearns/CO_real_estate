```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Files to Load 
income2015_to_load = "Resources/ACSST5Y2015_income_data.csv"
income2014_to_load = "Resources/ACSST5Y2014_income_data.csv"


# Read the Income Data
income2015_df = pd.read_csv(income2015_to_load)
income2014_df = pd.read_csv(income2014_to_load)
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(income2015_df, income2014_df, how="left", on=["GEO_ID", "GEO_ID"])

# Display the data table for preview
merged_df.head(5)
```
