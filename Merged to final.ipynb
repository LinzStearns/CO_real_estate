```python
# Dependencies and Setup
import pandas as pd


# Files to Load 
income2015_to_load = "Resources/income_2015.csv"
income2014_to_load = "Resources/income_2014.csv"
income2016_to_load = "Resources/income_2016.csv"
income2017_to_load = "Resources/income_2017.csv"
income2018_to_load = "Resources/income_2018.csv"

# Read the Income Data
income2015_df = pd.read_csv(income2015_to_load)
income2014_df = pd.read_csv(income2014_to_load)
income2016_df = pd.read_csv(income2016_to_load)
income2017_df = pd.read_csv(income2017_to_load)
income2018_df = pd.read_csv(income2018_to_load)
```

```python
R1415_df=pd.concat([income2015_df, income2014_df])
R141516_df=pd.concat([R1415_df, income2016_df])
R14151617_df=pd.concat([R141516_df, income2017_df])
R1415161718_df=pd.concat([R14151617_df, income2018_df])
R1415161718_df
```

```python
# Drop null values
R1415161718_df.dropna()
R1415161718_df
```

```python
# Export the CSV
R1415161718_df.to_csv('income_final.csv', index = False)
```
