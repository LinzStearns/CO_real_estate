```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Files to Load 
income2017_to_load = "Resources/ACSST5Y2017_income_data.csv"
income2016_to_load = "Resources/ACSST5Y2016_income_data.csv"


# Read the Income Data
income2017_df = pd.read_csv(income2017_to_load)
income2016_df = pd.read_csv(income2016_to_load)
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(income2017_df, income2016_df, how="left", on=["GEO_ID", "GEO_ID"])

# Display the data table for preview
merged_df.head(5)
```
