```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Files to Load 
income2019_to_load = "Resources/ACSST5Y2019_income_data.csv"
income2018_to_load = "Resources/ACSST5Y2018_income_data.csv"


# Read the Income Data
income2019_df = pd.read_csv(income2019_to_load)
income2018_df = pd.read_csv(income2018_to_load)
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(income2019_df, income2018_df, how="left", on=["GEO_ID", "GEO_ID"])

# Display the data table for preview
merged_df.head(5)
```
