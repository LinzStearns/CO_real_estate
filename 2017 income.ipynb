```python
# Import dependencies
import pandas as pd
from functools import reduce
```

```python
# Load in CSV as a DataFrame
income_df = pd.read_csv('Resources/ACSST5Y2017_income_data.csv')
income_df.head(10)
```

```python
# Drop null values
income_df.dropna()
```

```python
# Keep columns: income less than 50K
income_df50  = income_df.filter(['GEO_ID','Estimate!!Households!!Total','Estimate!!Households!!Total!!Less than $10,000', 'Estimate!!Households!!Total!!$10,000 to $14,999','Estimate!!Households!!Total!!$15,000 to $24,999', 'Estimate!!Households!!Total!!$25,000 to $29,999', 'Estimate!!Households!!Total!!$30,000 to $34,999', 'Estimate!!Households!!Total!!$35,000 to $39,999', 'Estimate!!Households!!Total!!$40,000 to $44,999', 'Estimate!!Households!!Total!!$45,000 to $49,999'])
income_df50.head()
```

```python
# Keep columns: income greater than 50K, less than 75K
income_df75  = income_df.filter(['GEO_ID','Estimate!!Households!!Total!!$50,000 to $74,999'])
income_df75
```

```python
# Keep columns: income greater than 75K, less than 100K
income_df_greaterthan75  = income_df.filter(['GEO_ID','Estimate!!Households!!Total!!$75,000 to $99,999'])
income_df_greaterthan75
```

```python
# Keep columns: income greater than 100K, less than 150K
income_df_greaterthan100  = income_df.filter(['GEO_ID','Estimate!!Households!!Total!!$100,000 to $149,999'])
income_df_greaterthan100
```

```python
# Keep columns: income greater than 150K, less than 200K
income_df_greaterthan150  = income_df.filter(['GEO_ID','Estimate!!Households!!Total!!$150,000 to $199,999'])
income_df_greaterthan150
```

```python
# Keep columns: income greater than 200K
income_df_greaterthan200  = income_df.filter(['GEO_ID','Estimate!!Households!!Total!!$200,000 or more'])
income_df_greaterthan200
```

```python
dfs = [income_df50, income_df75, income_df_greaterthan75, income_df_greaterthan100, income_df_greaterthan150, income_df_greaterthan200]
income_df = reduce(lambda left,right: pd.merge(left,right,on=['GEO_ID'],
                                            how='outer'), dfs)
income_df.head()
```

```python
# Export the CSV
income_df.to_csv('income_2017.csv', index = False)
```
