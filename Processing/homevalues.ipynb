```python
# Import dependencies
import pandas as pd
```

```python
# Load in CSV as a DataFrame
df = pd.read_csv('Resources/homevalues.csv')
display(df)
```

```python
# Drop null values
df.dropna()
```

```python
# Pull data for CO only
CO_homevalues_df = df[(df["StateName"] == "CO")]
CO_homevalues_df 
```

```python
# Remove unnecessary columns - RegionID, SizeRank, RegionType, State
CO_homevalues_df  = CO_homevalues_df.drop(['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)
CO_homevalues_df 
```

```python
# Pull data for Denver, CO only
CO_homevalues_df = df[(df["RegionName"] == "Denver, CO")]
CO_homevalues_df 
```

```python
# Export the CSV
CO_homevalues_df.to_csv('CO_homevalues.csv', index = False)
```
