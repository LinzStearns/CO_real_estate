```python
# Import dependencies 
import pandas as pd
```

```python
# File to Load
file_to_load = "Zillow-List-Sale-Price.csv"
```

```python
# Read the CSV
zillow_sale_df = pd.read_csv(file_to_load)
zillow_sale_df.head()
```

```python
# Drop null values
zillow_sale_df.dropna()
```

```python
# Pull data for CO only
CO_zillow_sale_df = zillow_sale_df[(zillow_sale_df["StateName"] == "CO")]
CO_zillow_sale_df                      
```

```python
# Remove unnecessary columns - RegionID, SizeRank, RegionType, State
CO_zillow_sale_df = CO_zillow_sale_df.drop(['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)
CO_zillow_sale_df
```

```python
# Export the CSV
CO_zillow_sale_df.to_csv(r'CO_zillow_sale.csv', index = False)
```
