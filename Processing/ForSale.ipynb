```python
# Import dependencies
import pandas as pd
```

```python
# Load in CSV as a DataFrame
df = pd.read_csv('Resources/forsale.csv')
display(df)
```

```python
# Drop null values
df.dropna()
```

```python
# Pull data for CO only
CO_forsale_df = df[(df["StateName"] == "CO")]
CO_forsale_df 
```

```python
# Remove unnecessary columns - RegionID, SizeRank, RegionType, State
CO_forsale_df  = CO_forsale_df.drop(['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)
CO_forsale_df 
```

```python
# Export the CSV
CO_forsale_df.to_csv('CO_forsale.csv', index = False)
```
